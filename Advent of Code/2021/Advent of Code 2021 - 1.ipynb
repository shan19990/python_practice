```python
with open("Advent of Code 2021 - 1.txt") as file:
    data = file.read()
```

```python
data_array = data.split('\n')
```

```python
count = 0
for i in range(len(data_array)-1):
    if data_array[i+1] > data_array[i]:
        print(data_array[i],data_array[i+1])
        count += 1
count
```

```python
count = 0
sum_array = []
for i in range(len(data_array)-3):
    sum = 0
    for j in range(3):
        sum += int(data_array[i+j])
    sum_array.append(sum)
for i in range(len(sum_array)-1):
    if sum_array[i+1]>sum_array[i]:
        count += 1
count
```
