```python
with open("Advent of Code 2021 - 2.txt") as file:
    data = file.read()
```

```python
data_list = data.split("\n")
```

```python
data_list
```

```python
x = 0
y = 0
for data in data_list:
    if data[0:7] == "forward":
        x = x + int(data[8:])
    elif data[0:4] == "down":
        y = y + int(data[5:])
    elif data[0:2] == "up":
        y = y - int(data[3:])
print(x*y)
```
