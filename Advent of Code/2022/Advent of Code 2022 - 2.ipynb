```python
with open("Advent of Code 2022 - 2.txt") as file:
    data = file.read()
```

```python
data = data.split("\n")
```

```python
Y = 2 #paper
X = 1 #rock
Z = 3 #scissor
score_dict = {"X":1,"Y":2,"Z":3}
```

```python
score = 0
turn = 0
for game in data:
    if game[0] == 'B':
        if game[2] == 'Y':
            turn = 3
        elif game[2] == 'X':
            turn = 0
        elif game[2] == 'Z':
            turn = 6
    elif game[0] == 'A':
        if game[2] == 'Y':
            turn = 6
        elif game[2] == 'X':
            turn = 3
        elif game[2] == 'Z':
            turn = 0
    elif game[0] == 'C':
        if game[2] == 'Y':
            turn = 0
        elif game[2] == 'X':
            turn = 6
        elif game[2] == 'Z':
            turn = 3
    score += score_dict[game[2]] + turn
print(score)
```

```python
score_dict = {"X":1,"Y":2,"Z":3}
```

X = 1 #rock
Y = 2 #paper
Z = 3 #scissor

A for Rock, B for Paper, and C for Scissors

X means you need to lose, Y means you need to end the round in a draw, and Z means you need to win.

```python
score = 0
for game in data:
    if game[0] == 'A':
        if game[2] == 'X':
            point = score_dict["Z"]
            turn = 0
        elif game[2] == 'Y':
            point = score_dict["X"]
            turn = 3
        elif game[2] == 'Z':
            point = score_dict["Y"]
            turn = 6
        #print(point + turn)
    elif game[0] == 'B':
        if game[2] == 'X':
            point = score_dict["X"]
            turn = 0
        elif game[2] == 'Y':
            point = score_dict["Y"]
            turn = 3
        elif game[2] == 'Z':
            point = score_dict["Z"]
            turn = 6
        #print(point + turn)
    elif game[0] == 'C':
        if game[2] == 'X':
            point = score_dict["Y"]
            turn = 0
        elif game[2] == 'Y':
            point = score_dict["Z"]
            turn = 3
        elif game[2] == 'Z':
            point = score_dict["X"]
            turn = 6
        #print(point + turn)
    score += point + turn
print(score)
```
