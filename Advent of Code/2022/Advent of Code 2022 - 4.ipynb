```python
with open("Advent of Code 2022 - 4.txt") as file:
    data = file.read()
```

```python
data_array = data.split("\n")
```

```python
data_array[0]
```

```python
first_worker = []
second_worker = []
similar_work = []
count = 0
for data in data_array:
    shifts = data.split(",")
    timings = shifts[0].split("-")
    for i in range(int(timings[0]),int(timings[1])+1):
        first_worker.append(i)
    #print(first_worker)
    first_worker_len = len(first_worker)
    #print(first_worker_len)
    timings = shifts[1].split("-")
    #print("-------------")
    for i in range(int(timings[0]),int(timings[1])+1):
        second_worker.append(i)
    #print(second_worker)
    second_worker_len = len(second_worker)
    #print(second_worker_len)
    #print("--------------------------------------------")
    for i in first_worker:
        if i in second_worker:
            similar_work.append(i)
    similar_work_len = len(similar_work)
    #print(similar_work)
    #print(similar_work_len)
    if similar_work_len == second_worker_len or similar_work_len == first_worker_len:
        count += 1
    first_worker = []
    second_worker = []
    similar_work = []
print(count)
```
