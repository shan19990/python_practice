```python
def check_if_greater(string,array,cube):
    number = ""
    #print(string)
    #print(array)
    flag = True
    for i in array:
        #print(string[i-3]+string[i-2])
        number = string[i-3] + string[i-2]
        if cube < int(number):
            flag = False
            break
    return flag
        
    
```

```python
def get_game_id(string):
    index = string.index(":")
    id = string[index-2]+string[index-1]
    return int(id)
```

```python
total = 0
for string in data:
    main_string = string

    green_index_list = []
    red_index_list = []
    blue_index_list = []
    start_index = 0

    while True:
        index = main_string.find("green", start_index)
        if index == -1:
            break
        green_index_list.append(index)
        start_index = index + 1
    start_index = 0
    while True:
        index = main_string.find("red", start_index)
        if index == -1:
            break
        red_index_list.append(index)
        start_index = index + 1
    start_index = 0    
    while True:
        index = main_string.find("blue", start_index)
        if index == -1:
            break
        blue_index_list.append(index)
        start_index = index + 1
    start_index = 0

    if check_if_greater(main_string,green_index_list,green_cube) and check_if_greater(main_string,red_index_list,red_cube) and check_if_greater(main_string,blue_index_list,blue_cube):
        total += get_game_id(main_string)
print(total)
```

```python
def check_which_lowest(string,array):
    number = ""
    #print(string)
    #print(array)
    max = 0
    for i in array:
        #print(string[i-3]+string[i-2])
        number = string[i-3] + string[i-2]
        if max < int(number):
            max = int(number)
    return max
        
    
```

```python
total = 0
for string in data:
    main_string = string

    green_index_list = []
    red_index_list = []
    blue_index_list = []
    start_index = 0

    while True:
        index = main_string.find("green", start_index)
        if index == -1:
            break
        green_index_list.append(index)
        start_index = index + 1
    start_index = 0
    while True:
        index = main_string.find("red", start_index)
        if index == -1:
            break
        red_index_list.append(index)
        start_index = index + 1
    start_index = 0    
    while True:
        index = main_string.find("blue", start_index)
        if index == -1:
            break
        blue_index_list.append(index)
        start_index = index + 1
    start_index = 0
    
    print(get_game_id(main_string))
    print(check_which_lowest(main_string,green_index_list) ,end =" ")
    print(check_which_lowest(main_string,red_index_list),end =" ")
    print(check_which_lowest(main_string,blue_index_list))
    
    
    power = check_which_lowest(main_string,green_index_list) * check_which_lowest(main_string,red_index_list) * check_which_lowest(main_string,blue_index_list)
    total += power
print(total)
```
