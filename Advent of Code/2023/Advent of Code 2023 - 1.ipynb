```python
def get_all_substrings(input_string):
    substrings = []
    n = len(input_string)

    for i in range(n):
        for j in range(i + 1, n + 1):
            substring = input_string[i:j]
            substrings.append(substring)

    return substrings

# Example usage:
input_str = "abc"
result = get_all_substrings(input_str)
print(result)
```
