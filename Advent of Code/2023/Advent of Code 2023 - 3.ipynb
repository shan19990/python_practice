```python
def is_symbol(char):
    return char in symbols
```

```python
symbols = ["&","*","/","@","#","=","!","$","-"]
```

```python
def cont_series(string):
    result = []
    number = ""
    for i in range(len(string)):
        if string[i].isdigit():
            number += string[i]
        elif (string[i-1]).isdigit() and not string[i].isdigit():
            number += ","
    array = number.split(",")
    array.pop()
    return array
```

```python
def if_symbol_is_next(data_array):
    for string in data_array:
        array = cont_series(string)
        for a in array:
            first_index = string.index(str(a))
            last_index = first_index+len(a)-1
            #print(string[first_index-1])
            if is_symbol(string[first_index-1]) or is_symbol(string[last_index+1]):
                print(a)
            #print(string[last_index+1])

        print("-------")
```

```python
if_symbol_is_next(data_array)
```
