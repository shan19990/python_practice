```python
import turtle as t
from time import strftime

# Function to draw clock hands
def draw_hand(length, angle):
    t.penup()
    t.goto(0, 0)
    t.setheading(angle)
    t.pendown()
    t.forward(length)
    t.penup()
    t.goto(0, 0)

# Set up the turtle screen
t.speed(1)
t.title("Clock Demo")

while True:
    t.clear()
    
    # Draw clock face
    t.circle(150)

    # Get current time
    current_time = strftime("%H:%M:%S")

    # Display current time
    t.write(current_time, align='center', font=('Arial', 14, 'normal'))

    # Draw clock hands
    draw_hand(100, int(current_time[:2]) % 12 * 30)
    draw_hand(130, int(current_time[3:5]) * 6)
    draw_hand(70, int(current_time[6:]) * 6)

    t.update()
    t.delay(1000)  # Update every second
```
