```python
import requests
from currency import access_key
```

```python
response = requests.get("http://api.currencylayer.com/list?access_key="+access_key.access_key)
cur_list = response.json()
print("The currency list is as follows:") 
for key in cur_list['currencies']:
    print(key)
```

```python
cur_from = input("Enter the currency you have: ")
cur_to = input("Enter the currency you want to have: ")
amount = input("Enter the amount to be converted: ")
url = "http://apilayer.net/api/live?access_key="+access_key.access_key+"&currencies="+cur_to+"&source="+cur_from
response = requests.get(url)
data = response.json()
rate = data['quotes'][cur_from+cur_to]
curr_after_convert = int(amount) * rate
print(f"{amount} converted from {cur_from} to {cur_to} comes out to {curr_after_convert}")
```
